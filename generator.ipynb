# AFCON Sweepstake

Generate team selection for participants.

Assumes 12 participants & 24 teams.

RNG seeded for reproducibility.

Attempts to copy generated teams to clipboard.

```python
import pandas as pd
from numpy.random import default_rng

rng = default_rng(51)
```

```python
participants = [
    "Daniel",
    "Lewis",
    "Liam",
    "Mark",
    "Neil",
    "Ollie",
    "Sam",
    "Stephen",
    "Stevie",
    "Struan",
    "Thomas",
    "Toby",
]

teams = [
    "Algeria",
    "Burkina Faso",
    "Cameroon",
    "Cape Verde",
    "Comoros",
    "Egypt",
    "Equatorial Guinea",
    "Ethiopia",
    "Gabon",
    "Gambia",
    "Ghana",
    "Guinea",
    "Guinea-Bissau",
    "Ivory Coast",
    "Malawi",
    "Mali",
    "Mauritania",
    "Morocco",
    "Nigeria",
    "Senegal",
    "Sierra Leone",
    "Sudan",
    "Tunisia",
    "Zimbabe",
]

assert participants == sorted(participants)
assert teams == sorted(teams)
```

```python
sweepstake = pd.DataFrame(
    index=participants,
    data=rng.choice(teams, size=len(teams), replace=False).reshape(-1, 2),
    columns=["Team 1", "Team 2"],
)

sweepstake.to_clipboard()
sweepstake
```
